```
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
```

```
def update_metadata(metadata_table_path,dataset_name_for_reset,dataset_path_for_reset,reset_date_for_dataset):
  print("updating metadata")
  
  def _get_spark():
    spark = SparkSession.builder.appName("project_customer_360_data_flow_control").getOrCreate()
    spark.conf.set("spark.sql.sources.partitionOverwriteMode", "dynamic")
    spark.conf.set("spark.sql.parquet.binaryAsString", "true")
    spark.conf.set("spark.sql.shuffle.partitions", 200)
    spark.conf.set("spark.sql.files.maxPartitionBytes", 1024*1024*256)
    return spark
  spark = _get_spark()
  spark.conf.set("spark.sql.sources.partitionOverwriteMode", "dynamic")

  
  metadata_table_update_df = spark.range(1)

  metadata_table_update_df = (
            metadata_table_update_df.withColumn("table_name", F.lit(dataset_name_for_reset))
                .withColumn("table_path", F.lit(dataset_path_for_reset))
                .withColumn("write_mode", F.lit("overwrite"))
                .withColumn("target_max_data_load_date", F.to_date(F.lit(reset_date_for_dataset), "yyyy-MM-dd"))
                .withColumn("updated_on", F.current_date())
                .drop("id")
                    )

  metadata_table_update_df.write.partitionBy("table_name").format("parquet").mode("overwrite").save(metadata_table_path)
  
```

```
def create_table(obj): 
  final_str = ''
  list_of_item = []
  try:
    print("Reading Data -->> {0}".format(obj['path']))
    df = spark.read.parquet(obj['path'])
  except Exception as e:
    df=None
    print("No Path Present for {} {}".format(obj['table_name'], obj['path']))
  if df:
    metadata_table_path = '/mnt/customer360-blob-output/C360/UTILITIES/metadata_table/'
    dataset_name_for_reset = obj['table_name']
    dataset_path_for_reset = obj['path']
    if 'l1_features' in obj['path']:
      reset_date_for_dataset = df.select(F.max(F.col("event_partition_date")).alias("max_date")).collect()[0].max_date
      update_metadata(metadata_table_path,dataset_name_for_reset,dataset_path_for_reset,reset_date_for_dataset)
      
    elif 'l2_features' in obj['path']:
      reset_date_for_dataset = df.select(F.max(F.col("start_of_week")).alias("max_date")).collect()[0].max_date
      update_metadata(metadata_table_path,dataset_name_for_reset,dataset_path_for_reset,reset_date_for_dataset)
      
    elif 'l3_features' in obj['path']:
      reset_date_for_dataset = df.select(F.max(F.col("start_of_month")).alias("max_date")).collect()[0].max_date
      update_metadata(metadata_table_path,dataset_name_for_reset,dataset_path_for_reset,reset_date_for_dataset)
      
    elif 'l4_features' in obj['path'] and 'event_partition_date' in df.columns:
      reset_date_for_dataset = df.select(F.max(F.col("event_partition_date")).alias("max_date")).collect()[0].max_date
      update_metadata(metadata_table_path,dataset_name_for_reset,dataset_path_for_reset,reset_date_for_dataset)
      
    elif 'l4_features' in obj['path'] and 'start_of_week' in df.columns:
      reset_date_for_dataset = df.select(F.max(F.col("start_of_week")).alias("max_date")).collect()[0].max_date
      update_metadata(metadata_table_path,dataset_name_for_reset,dataset_path_for_reset,reset_date_for_dataset)
      
    elif 'l4_features' in obj['path'] and 'start_of_month' in df.columns:
      reset_date_for_dataset = df.select(F.max(F.col("start_of_month")).alias("max_date")).collect()[0].max_date
      update_metadata(metadata_table_path,dataset_name_for_reset,dataset_path_for_reset,reset_date_for_dataset)
      
    else:
      print("No partition column found")
```

```
for curr_item in metadata_on_cloud['elements']:
  create_table(curr_item)
```
