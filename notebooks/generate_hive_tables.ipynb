```python
def create_table(obj):
    final_str = ''
    list_of_item = []
    try:
        df = spark.read.parquet(obj['path'])
    except Exception as e:
        df=None
        print("No Path Present for {} {}".format(obj['table_name'], obj['path']))
    if df:
        for curr_col in df.schema.jsonValue()['fields']:
            temp_str = curr_col['name']+' '+curr_col['type']
            list_of_item.append(temp_str)
        final_str =  ", ".join(list_of_item)
        drop_str = "drop table if exists {}.{}".format(database_name, obj['table_name'])
        print("Dropping Table {}.{}".format(database_name, obj['table_name']))
        spark.sql(drop_str)
        print("Creating Table {} from path {}".format(obj['table_name'], obj['path']))                
        stmt = "create table if not exists {}.{} ({}) using parquet LOCATION '{}'".format(database_name, obj['table_name'], final_str, obj['path'])
        spark.sql(stmt)
```

```python
# on cloud
for curr_item in metadata_on_cloud['elements']:
  create_table(curr_item)
```

```python
# on premise
for curr_item in metadata_on_premise['elements']:
  create_table(curr_item)
```
